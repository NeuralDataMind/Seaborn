**Value Retention**

- Filter your data down to the following values of make (brand): ['Ford', 'BMW', 'Toyota', 'Chevy']
- Build alinear relatonship plot comparing odmeter (distance car has travelled) vs. sellingprice, broken out by make.
- If you want to fit a regression with scipy.stats, use dropna() on the dataframe first

```python
brand_list = ['Ford', 'BMW', 'Toyota', 'Chevrolet']

sns.lmplot(
    x="odometer",
    y="sellingprice",
    data=cars.query("make in @brand_list and sellingprice < 150000 and odometer < 200000").sample(1000),
    hue="make",
    scatter_kws={"alpha": .3}
)
```

```python
import scipy.stats as stats

cars_regress = cars.dropna()

stats.linregress(x=cars_regress["odometer"], y=cars_regress["sellingprice"])
```

```python
for brand in brand_list:
    data = cars_regress.loc[cars_regress["make"] == brand]
    print(f"{brand}: {stats.linregress(data['odometer'], data['sellingprice'])[0:2]}")
```

**Categorical Variable Relationships**

- Bin the condition variable into increments of .5, For example, the first bin will be '1 - 1.5', the second '1.5 - 2' and on up until the maximun condition value of 5, Use pd.cut() to create the bins.
- Build a barplot of the average price by condition bin.

```python
fig, ax = plt.subplots(figsize=(10, 10))

bins = np.linspace(1, 5, 9)
labels = ["1-1.5", "1.5-2", "2-2.5", "2.5-3", "3-3.5", "3.5-4", "4-4.5", "4.5-5"]

cars["condition_bins"] = pd.cut(cars["condition"], bins=bins, labels=labels)

sns.barplot(
    x="condition_bins",
    y="sellingprice",
    data=cars,
    palette="husl"
)

ax.set_title("Average Price by Condition")

sns.despine()
```

**Price by make and condition**

- Create a pivot table with the top 10 most common car brands make column as rows, and condition bins as columns. The values in the table shpuld be the average selling price for each make/condition combination. Consider using to get the top 10 most common brands.
- Once you've created the table, build a heatmap from the data. Which brands and quality level have the highest selling prices? Does this this match your expecatons?

```python
cars["make"].value_counts()
```

```python
top10 = cars["make"].value_counts().iloc[:10].index

avgs = (cars
        .loc[cars["make"].isin(top10)]
        .pivot_table(index="make",
                     columns="condition_bins",
                     values={"sellingprice": "mean"}
                     )
        .droplevel(0, axis=1) 
        )

avgs.head()
```

```python
fig, ax = plt.subplots(figsize=(10,10))

plt.title("Avg. Selling Price by Make and Quality", fontsize=16)

sns.heatmap(
    avgs,
    annot=True,
    cmap="RdYlGn",
    fmt="g"
)
```

**Are there any "Deals"?**

Ok, thanks to the chart above we are getting closer to honing in on the truck to purchase.

- Create a cloumn diff_to_mmr which is the difference between sellingprice and mmr.
- Then build a barplot looking at mean diff_to_mmr by color.
- Which color sells for more than recommended price?

```python
sns.set_style("darkgrid")

f150s = f150s.assign(diff_to_mmr = f150s["sellingprice"] - f150s["mmr"])

f150s_by_color = (
    f150s
    .groupby("color", as_index=False)
    .agg({"diff_to_mmr": "mean"})
)

fig, ax = plt.subplots(figsize = (8, 6))

sns.barplot(
    x="color",
    y="diff_to_mmr",
    data=f150s_by_color,
    ci=None,
    palette="husl"
)

plt.xticks(rotation=45)

plt.show()
```

```python
f150s.head()
```

```python
trim_list =['XL', 'XLT', 'Platinum', 'svt Raptor']

f150s_reduced = f150s.query("color != 'orange' and trim in @trim_list")

f150s_pivot = (
    f150s_reduced
    .loc[f150s_reduced["year"].between(2009, 2014)]
    .pivot_table(
        index="trim",
        columns="condition_bins",
        values={"sellingprice": "mean"}
    )
)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

sns.heatmap(f150s_pivot, annot=True, fmt="g")
```

**Best state to buy truck n?**

- Ok, we've decided on the XLT model, it has more features than XL but is still quite affordable. Also filter out any truck with a quity lesss than 3.5. Your DataFrame should only have truck with qualitly of 3.5 or greater.
- Build asubplot with a bar chart of state by diff_to_mmr and state by count to find which states sell XLT models below mmr and have ample quantity

```python
f150s_XLT = (f150s_reduced
             .query("trim == 'XLT' and condition >= 3.5")
             .groupby(['state'], as_index=False)
             .agg(
                 mean_diff=("diff_to_mmr", 'mean'),
                 auto_count=("diff_to_mmr", 'count')
             )
)

fig, ax = plt.subplots(2, figsize=(8, 6))

sns.barplot(
    x="state",
    y="mean_diff",
    data = f150s_XLT,
    ax=ax[0]
)

sns.barplot(
    x="state",
    y="auto_count",
    data = f150s_XLT,
    ax=ax[1]

)
```
